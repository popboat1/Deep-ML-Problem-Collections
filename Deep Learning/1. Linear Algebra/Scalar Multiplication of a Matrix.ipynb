## Scalar Multiplication of a Matrix

When a matrix $A$ is multiplied by a scalar $k$, the operation is defined as multiplying **each element** of $A$ by $k$.

### Given a Matrix $A$

$$
A =
\begin{pmatrix}
a_{11} & a_{12} \\
a_{21} & a_{22}
\end{pmatrix}
$$

### Scalar Multiplication

Given a scalar $k$, the result of the scalar multiplication $kA$ is:

$$
kA =
\begin{pmatrix}
k a_{11} & k a_{12} \\
k a_{21} & k a_{22}
\end{pmatrix}
$$

### Interpretation

This operation **scales the matrix by $k$** without changing:

- The **dimension** of the matrix
- The **relative proportion** between its elements

[Problem Desc](https://www.deep-ml.com/problems/5)

```python
def scalar_multiply(matrix: list[list[int|float]], scalar: int|float) -> list[list[int | float]]:
    return [[matrix[j][i] * scalar for i in range(len(matrix[0]))] for j in range(len(matrix))]
```

```python
# Test Case 1
print(scalar_multiply([[1,2],[3,4]], 2))

# Expected Output:
# [[2, 4], [6, 8]]
```

```python
# Test Case 2
print(scalar_multiply([[0,-1],[1,0]], -1))

# Expected Output:
# [[0, 1], [-1, 0]]
```
