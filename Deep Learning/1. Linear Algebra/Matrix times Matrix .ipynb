## Pure Python

```python
def matrixmul(a: list[list[int|float]], b: list[list[int|float]]) -> list[list[int|float]]:
    if len(a[0]) != len(b):
        return -1
    
    return [[sum(x * y for x, y in zip(row_a, col_b)) for col_b in zip(*b)] for row_a in a]
```

### Numpy

```python
import numpy as np

def matrixmul(a: list[list[int|float]], b: list[list[int|float]]) -> list[list[int|float]]:
    c = (np.array(a) @ np.array(b)).tolist()
    return c
```

```python
# Test Case 1
print(matrixmul([[1,2,3],[2,3,4],[5,6,7]],[[3,2,1],[4,3,2],[5,4,3]]))

# Expected Output:
# [[26, 20, 14], [38, 29, 20], [74, 56, 38]]
```

```python
# Test Case 2
print(matrixmul([[0,0],[2,4],[1,2]],[[0,0],[2,4]]))

# Expected Output:
# [[0, 0], [8, 16], [4, 8]]
```
