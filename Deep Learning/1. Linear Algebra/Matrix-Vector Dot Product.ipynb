## Matrix–Vector Dot Product

Consider a matrix $A$ and a vector $v$.

### Matrix $A$ ($n \times m$)

$$
A =
\begin{pmatrix}
a_{11} & a_{12} & \dots & a_{1m} \\
a_{21} & a_{22} & \dots & a_{2m} \\
\vdots & \vdots & \ddots & \vdots \\
a_{n1} & a_{n2} & \dots & a_{nm}
\end{pmatrix}
$$

### Vector $v$ (length $m$)

$$
v =
\begin{pmatrix}
v_1 \\
v_2 \\
\vdots \\
v_m
\end{pmatrix}
$$

### Matrix–Vector Dot Product

The dot product $A \cdot v$ produces a new vector of length $n$:

$$
A \cdot v =
\begin{pmatrix}
a_{11}v_1 + a_{12}v_2 + \dots + a_{1m}v_m \\
a_{21}v_1 + a_{22}v_2 + \dots + a_{2m}v_m \\
\vdots \\
a_{n1}v_1 + a_{n2}v_2 + \dots + a_{nm}v_m
\end{pmatrix}
$$

### Key Requirement

The number of columns in the matrix ($m$) must equal the length of the vector ($m$).

If this condition is not satisfied, the operation is **undefined**, and the function should return **`-1`**.

[Problem Desc](https://www.deep-ml.com/problems/1)

```python
def matrix_dot_vector(a: list[list[int|float]], b: list[int|float]) -> list[int|float]:
    # Return a list where each element is the dot product of a row of 'a' with 'b'
    # If the number of columns in 'a' doesn't match the length of 'b', return -1
    return [sum(x * y for x, y in zip(row, b)) for row in a]
```

```python
# Test Case 1
print(matrix_dot_vector([[1, 2, 3], [2, 4, 5], [6, 8, 9]], [1, 2, 3]))

# Expected Output:
# [14, 25, 49]
```

```python
# Test Case 2
print(matrix_dot_vector([[1, 2], [2, 4], [6, 8], [12, 4]], [1, 2, 3]))

# Expected Output:
# -1
```
