## Calculate Mean by Row or Column

Calculating the **mean of a matrix by row or column** involves averaging the elements across a specified dimension. This operation provides insights into the distribution of values within a dataset and is commonly used in **data normalization** and **feature scaling**.

---

## Row Mean

The mean of a row is computed by summing all elements in the row and dividing by the number of elements.

For row $i$, the mean is:

$$
\mu_{\text{row } i} = \frac{1}{n} \sum_{j=1}^{n} a_{ij}
$$

where:

- $a_{ij}$ is the matrix element in the $i$-th row and $j$-th column  
- $n$ is the total number of columns  

---

## Column Mean

Similarly, the mean of a column is calculated by summing all elements in the column and dividing by the number of elements.

For column $j$, the mean is:

$$
\mu_{\text{column } j} = \frac{1}{m} \sum_{i=1}^{m} a_{ij}
$$

where:

- $a_{ij}$ is the matrix element in the $i$-th row and $j$-th column  
- $m$ is the total number of rows  

---

## Interpretation

This formulation helps us understand how data is **aggregated across different dimensions** of a matrix. Row and column means are fundamental operations used in many **data preprocessing techniques**, such as:

- Feature scaling  
- Data normalization  
- Centering data before performing methods like **PCA** or **SVD**

```python
def calculate_matrix_mean(matrix: list[list[float|int]], mode: str) -> list[float]:
    means = []
    if mode == 'column':
        new = [[matrix[i][j] for i in range(len(matrix))] for j in range(len(matrix[0]))]
        for i in new:
            res = 0
            for j in i:
                res += j
            means.append(res / len(matrix))
    elif mode == 'row':
        for i in matrix:
            res = 0
            for j in i:
                res += j
            means.append(res / len(matrix[0]))
    
    return means
```

### Faster

```python
def calculate_matrix_mean(matrix: list[list[float|int]], mode: str) -> list[float]:
    if mode == 'row':
        return [sum(row) / len(row) for row in matrix]
    elif mode == 'column':
        return [sum(col) / len(matrix) for col in zip(*matrix)]
```

```python
# Test Case 1
print(calculate_matrix_mean([[1, 2, 3], [4, 5, 6], [7, 8, 9]], 'column'))

# Expected Output:
# [4.0, 5.0, 6.0]
```

```python
# Test Case 2
print(calculate_matrix_mean([[1, 2, 3], [4, 5, 6], [7, 8, 9]], 'row'))

# Expected Output:
# [2.0, 5.0, 8.0]
```
