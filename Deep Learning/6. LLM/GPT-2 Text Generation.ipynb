# Understanding Transformer Architecture and Text Generation

Transformers have revolutionized the field of Natural Language Processing (NLP) with their efficient and scalable architecture. This guide provides an in-depth look into the core components of transformers and how they facilitate advanced text generation.

# 1. Introduction to Transformers

Transformers are a groundbreaking neural network architecture that has significantly advanced NLP. Introduced in the seminal paper **"Attention is All You Need"** by Vaswani et al. (2017), transformers have outperformed traditional models like **Recurrent Neural Networks (RNNs)** and **Long Short-Term Memory networks (LSTMs)** in various NLP tasks.

## Key Advantages of Transformers

### Parallel Processing
Unlike RNNs, which process input sequences sequentially, transformers handle entire sequences simultaneously. This parallelism leads to substantial improvements in training speed and efficiency.

### Scalability
Transformers can effectively scale to handle large datasets and complex tasks, making them ideal for applications like language translation, text generation, and summarization.

### Self-Attention Mechanism
The core innovation of transformers is the **self-attention mechanism**, which allows the model to weigh the importance of different words in a sequence relative to each other. This capability enables the model to capture long-range dependencies and contextual relationships within the text.

## Applications of Transformers

- **Text Generation** – Creating coherent and contextually relevant text based on a given prompt  
- **Machine Translation** – Translating text from one language to another with high accuracy  
- **Text Summarization** – Condensing long documents into concise summaries while retaining key information  
- **Question Answering** – Providing accurate answers to user queries based on contextual understanding

# 2. Core Concepts

To fully grasp the transformer architecture, it's essential to understand its foundational components. Below are the core concepts that constitute the building blocks of transformers.

---

# 2.1 GELU Activation Function

The **Gaussian Error Linear Unit (GELU)** is an advanced activation function that enhances the performance of deep neural networks.

## Mathematical Expression
$$GELU(x) = 0.5 * x * (1 + tanh( sqrt(\frac{2}{π}) * (x + 0.044715 * x^3) ))$$


## Purpose

GELU introduces non-linearity in the network while maintaining smooth gradient flow. Unlike **ReLU** or **Sigmoid**, GELU provides a probabilistic activation behavior, improving performance in deep architectures.

## Benefits

- **Smooth Activation** – Reduces the likelihood of dead neurons compared to ReLU  
- **Improved Gradient Flow** – Helps stabilize training and prevents vanishing or exploding gradients  

---

# 2.2 Softmax for Attention

Softmax converts raw attention scores into a probability distribution so that the weights sum to one.

## Mathematical Expression

$$Softmax(x_i) = \frac{exp(x_i)}{\sum_{j=1}^{N} exp(x_j)}$$


## Purpose

Within attention mechanisms, softmax normalizes attention scores so the model can focus more on important tokens.

## Example

If attention scores are:
[2, 1, 0.1]


After softmax:
[0.659, 0.242, 0.099]


This indicates the relative importance of each token.

---

# 2.3 Layer Normalization

Layer normalization stabilizes and accelerates training by normalizing inputs across features.

## Mathematical Expression

$$LayerNorm(x) = g \cdot \frac{(x - μ)}{sqrt(σ² + ε)} + b$$


Where:

- **μ** – mean of x along the last axis  
- **σ²** – variance of x  
- **g, b** – learnable scale and bias parameters  
- **ε** – small constant preventing division by zero  

## Purpose

Ensures each layer receives inputs with consistent distributions, improving convergence and training stability.

---

# 2.4 Multi-Head Attention

Multi-head attention allows the model to focus on different representation subspaces simultaneously.

## Components

### Query (Q), Key (K), Value (V)

Each attention head projects the input embeddings into Q, K, and V matrices.

### Scaled Dot-Product Attention
$$Attention(Q, K, V) = Softmax((\frac{QK^T}{sqrt(d_k)}) V$$


Where:

- **d_k** is the dimensionality of the key vectors.

### Important Note

The embedding dimension must be divisible by the number of heads.

Example:

Embedding dimension = 10
Number of heads = 2
Dimensions per head = 5


## Benefits

- **Diverse Attention Patterns** – Multiple heads focus on different relationships  
- **Better Feature Representation** – Combines multiple perspectives of the input  

---

# 2.5 Feedforward Network (FFN)

The Feedforward Network processes each token independently after attention.

## Mathematical Expression
$$FFN(x) = Linear2(GELU(Linear1(x)))$$


## Structure

1. **Linear Layer 1** – Expands dimensionality  
2. **GELU Activation** – Adds non-linearity  
3. **Linear Layer 2** – Projects back to original dimension  

## Purpose

Enhances the model's ability to learn complex representations.

---

# 2.6 Transformer Block

A **Transformer Block** combines attention and feedforward layers with residual connections and normalization.

## Structure

### Multi-Head Attention Layer
$$x1 = LayerNorm(x + MHA(x))$$


### Feedforward Network
$$x2 = LayerNorm(x1 + FFN(x1))$$


## Advantages

- **Supports Deep Architectures**  
- **Residual Connections improve gradient flow**  
- **Modular design allows stacking multiple blocks**

---

# 2.7 GPT-2 Text Generation

**GPT-2 (Generative Pre-trained Transformer 2)** uses transformer blocks for autoregressive text generation.

## Key Components

- **Token Embeddings** – Represent token meaning  
- **Positional Embeddings** – Encode token position  
- **Causal Attention** – Prevents attending to future tokens  
- **Stacked Transformer Blocks** – Iteratively refine representations  

## Text Generation Process

1. Provide a prompt.
2. Convert the prompt into tokens.
3. Convert tokens to embeddings.
4. Process embeddings through transformer blocks.
5. Predict the next token probability.
6. Append the predicted token to the sequence.
7. Repeat until the desired length is reached.

Token embeddings map input tokens to **dense vector representations**.

Each token in the vocabulary is converted into a vector so that the model can process it numerically.

### 2. Positional Embeddings

Positional embeddings add **position information** to token embeddings.

Since transformer models process sequences in parallel, positional embeddings help the model understand the **order of tokens** in the sequence.

### 3. Multi-head Attention

Multi-head attention allows the model to **attend to multiple parts of the sequence simultaneously**.

This mechanism enables the model to capture relationships between tokens at different positions.

### 4. Feed-Forward Network

The feed-forward network processes the outputs from the attention mechanism through dense layers.

It applies a **non-linear transformation** to enhance the representation learned from attention.

### 5. Layer Normalization

Layer normalization stabilizes the learning process by normalizing the activations within a layer.

This helps improve **training stability and convergence**.

````python
import numpy as np

def gen_text(prompt: str, n_tokens_to_generate: int = 40):
    encoder, hparams, params = load_encoder_hparams_and_params()
    
    tokens = encoder.encode(prompt)
    prompt_length = len(tokens)
    
    for i in range(n_tokens_to_generate):
        token_embeddings = params["wte"][tokens]
        pos_embeddings = params["wpe"][range(len(tokens))]
        
        x = token_embeddings + pos_embeddings
        
        x_last = x[-1]
        
        mean = np.mean(x_last)
        var = np.var(x_last)
        
        normalized = (x_last - mean) / np.sqrt(var + 1e-5) * params["ln_f"]["g"] + params["ln_f"]["b"]
        
        logits = normalized @ params["wte"].T
        
        next_token = np.argmax(logits)
        
        tokens.append(next_token)
    
    return encoder.decode(tokens[prompt_length:])
        

def load_encoder_hparams_and_params(model_size: str = "124M", models_dir: str = "models"):
    class DummyBPE:
        def __init__(self):
            self.encoder_dict = {"hello": 1, "world": 2, "<UNK>": 0}
            
        def encode(self, text: str):
            tokens = text.strip().split()
            return [self.encoder_dict.get(token, self.encoder_dict["<UNK>"]) for token in tokens]
        
        def decode(self, token_ids: list):
            reversed_dict = {v: k for k, v in self.encoder_dict.items()}
            return " ".join([reversed_dict.get(tok_id, "<UNK>") for tok_id in token_ids])
        
    hparams = {
		"n_ctx": 1024,
		"n_head": 2
	}
    
    params = {
		"wte": np.random.rand(3, 10),
		"wpe": np.random.rand(1024, 10),
		"blocks": [],
		"ln_f": {
			"g": np.ones(10),
			"b": np.zeros(10),
		}
	}
    
    encoder = DummyBPE()
    return encoder, hparams, params
````
