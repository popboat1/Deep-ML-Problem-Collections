## Understanding Descriptive Statistics

Descriptive statistics provide a summary of data through various measures that help understand the basic structure and distribution of data.

---

## Key Metrics

### Central Tendency

- **Mean**: The average of all values:

  $$
  \bar{x} = \frac{\sum_{i=1}^{n} x_i}{n}
  $$

- **Median**: The middle value when the data is sorted.  
  For even-length datasets, it is the average of the two middle values.

- **Mode**: The most frequently occurring value.  
  If multiple values share the highest frequency, return the **smallest**.

---

### Spread / Dispersion

- **Population Variance**: Measures how spread out values are from the mean.

  $$
  \sigma^2 = \frac{\sum_{i=1}^{n}(x_i - \bar{x})^2}{n}
  $$

  **Note:** This uses **$n$ (population variance)**, not **$n-1$ (sample variance)**.

- **Standard Deviation**: The square root of variance.

  $$
  \sigma = \sqrt{\sigma^2}
  $$

---

### Distribution

- **Percentiles**: Values below which a percentage of the data falls.

  - **25th percentile (Q1)**: 25% of data is below this value  
  - **50th percentile (Q2)**: Same as the **median**  
  - **75th percentile (Q3)**: 75% of data is below this value  

- **Interquartile Range (IQR)**: Measures the spread of the middle 50% of the data.

  $$
  IQR = Q3 - Q1
  $$

---

## Example Calculation

For the dataset:
[1, 2, 2, 3, 4, 4, 4, 5]


| Metric | Calculation | Result |
|------|------|------|
| Mean | (1+2+2+3+4+4+4+5)/8 | 3.125 |
| Median | (3+4)/2 | 3.5 |
| Mode | 4 appears 3 times | 4 |
| Variance | $\sum (x_i - 3.125)^2 / 8$ | 1.6094 |
| Std Dev | $\sqrt{1.6094}$ | 1.2686 |
| IQR | $Q3 - Q1 = 4.0 - 2.0$ | 2.0 |

---

## Applications

- **Exploratory Data Analysis (EDA)**: Initial data understanding  
- **Outlier Detection**: Using IQR (values outside $Q1 - 1.5 \times IQR$ to $Q3 + 1.5 \times IQR$)  
- **Feature Engineering**: Normalizing and scaling features  
- **Data Quality Analysis**: Detecting anomalies in distributions

```python
import numpy as np

def descriptive_statistics(data: list | np.ndarray) -> dict:
    """
    Calculate various descriptive statistics metrics for a given dataset.
    
    Args:
        data: List or numpy array of numerical values
    
    Returns:
        Dictionary containing mean, median, mode, variance, standard deviation,
        percentiles (25th, 50th, 75th), and interquartile range (IQR)
    """
    X = np.array(data)
    mean = np.mean(X)
    median = np.median(X)
    
    unique_values, counts = np.unique(data, return_counts=True)
    
    max_count_index = np.argmax(counts)
    mode = unique_values[max_count_index]
    
    variance = np.var(X)
    std = np.std(X)
    percentile_25 = np.percentile(X, 25)
    percentile_50 = np.percentile(X, 50)
    percentile_75 = np.percentile(X, 75)
    
    IQR = percentile_75 - percentile_25
    
    return {
		'mean': mean,
		'median': median,
		'mode': mode,
		'variance': variance,
		'standard_deviation': std,
		'25th_percentile': percentile_25,
		'50th_percentile': percentile_50,
		'75th_percentile': percentile_75,
		'interquartile_range': IQR
	}
```

```python
# Test Case 1
result = descriptive_statistics([1, 2, 2, 3, 4, 4, 4, 5])
print({k: round(v, 4) if isinstance(v, float) else v for k, v in result.items()})

# Expected Output:
# {'mean': 3.125, 'median': 3.5, 'mode': 4, 'variance': 1.6094, 'standard_deviation': 1.2686, '25th_percentile': 2.0, '50th_percentile': 3.5, '75th_percentile': 4.0, 'interquartile_range': 2.0}
```
