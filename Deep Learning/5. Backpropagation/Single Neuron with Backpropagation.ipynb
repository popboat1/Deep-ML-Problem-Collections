# Neural Network Learning with Backpropagation

This task involves implementing backpropagation for a single neuron in a neural network. The neuron processes inputs and updates parameters to minimize the Mean Squared Error (MSE) between predicted outputs and true labels.

---

## Mathematical Background

### Forward Pass


Compute the neuron output by calculating the dot product of the weights and input features, and adding the bias:

$$
z = w_1x_1 + w_2x_2 + \dots + w_nx_n + b
$$

Then, apply the activation function (in this case, the sigmoid function):

$$
\sigma(z) = \frac{1}{1 + e^{-z}}
$$

### Loss Calculation (MSE)
The Mean Squared Error quantifies the error between the neuron's predictions and the actual labels:

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^n (\sigma(z_i) - y_i)^2
$$

*(Where $n$ is the number of samples, $\sigma(z_i)$ is the predicted output, and $y_i$ is the true label).*

---

### Backward Pass (Gradient Calculation)


Compute the gradient of the MSE with respect to each weight and the bias. This involves the partial derivatives of the loss function with respect to the output of the neuron, multiplied by the derivative of the sigmoid function (using the Chain Rule):

**Gradient with respect to weights ($w_j$):**
$$
\frac{\partial \text{MSE}}{\partial w_j} = \frac{2}{n} \sum_{i=1}^n (\sigma(z_i) - y_i)\sigma'(z_i)x_{ij}
$$

**Gradient with respect to bias ($b$):**
$$
\frac{\partial \text{MSE}}{\partial b} = \frac{2}{n} \sum_{i=1}^n (\sigma(z_i) - y_i)\sigma'(z_i)
$$

---

### Parameter Update


Update each weight and the bias by subtracting a portion of the gradient, determined by the learning rate ($\alpha$):

**Update weights:**
$$
w_j = w_j - \alpha \frac{\partial \text{MSE}}{\partial w_j}
$$

**Update bias:**
$$
b = b - \alpha \frac{\partial \text{MSE}}{\partial b}
$$

---

## Practical Implementation
This process refines the neuron's ability to predict accurately by iteratively adjusting the weights and bias based on the error gradients, optimizing the neural network's performance over multiple iterations (epochs).

[Problem Desc](#https://www.deep-ml.com/problems/25)

```python
import numpy as np

def train_neuron(features: np.ndarray, labels: np.ndarray, initial_weights: np.ndarray, initial_bias: float, learning_rate: float, epochs: int) -> (np.ndarray, float, list[float]):
    N = len(features)
    weights = initial_weights.copy()
    bias = initial_bias
    mse_values = []
    
    for epoch in range(epochs):
        #1. forward pass
        z = np.dot(features, weights) + bias
        a = 1 / (1 + np.exp(-z))
        
        #2. calc loss
        mse1 = (a - labels) ** 2
        mse = 1/N * np.sum(mse1)
        mse_values.append(round(mse, 4))
        
        #3. backward pass
        dz = 2/N * (a - labels) * a * (1 - a)
        
        dW = np.dot(features.T, dz)
        db = np.sum(dz)
        
        #4. update weight and bias
        weights -= learning_rate * dW
        bias -= learning_rate * db
    
    updated_weights = np.round(weights, 4).tolist()
    updated_bias = float(np.round(bias, 4))
    
    return updated_weights, updated_bias, mse_values
```

```python
# Test Case 1
print(train_neuron(np.array([[1.0, 2.0], [2.0, 1.0], [-1.0, -2.0]]), np.array([1, 0, 0]), np.array([0.1, -0.2]), 0.0, 0.1, 2))

# Expected Output:
# ([0.1036, -0.1425], -0.0167, [0.3033, 0.2942])
```

```python
# Test Case 2
print(train_neuron(np.array([[1, 2], [2, 3], [3, 1]]), np.array([1, 0, 1]), np.array([0.5, -0.2]), 0, 0.1, 3))

# Expected Output:
# ([0.4892, -0.2301], 0.0029, [0.21, 0.2087, 0.2076])
```
