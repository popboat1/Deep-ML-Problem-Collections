# Understanding Gradient Descent Variants with MSE Loss

Gradient Descent is an optimization algorithm used to minimize the cost function in machine learning models. The Mean Squared Error (MSE) loss function is commonly used in regression tasks. There are three main variants based on how much data is used per weight update:

## Terminology
* **Epoch:** One complete pass through the entire training dataset.
* **Iteration/Update:** One weight update step.



---

## 1. Batch Gradient Descent
Computes the gradient using the **entire dataset** before making one weight update. 
* *One epoch = one weight update.*

$$
\theta = \theta - \alpha \cdot \frac{2}{m} \sum_{i=1}^m (h_\theta(x^{(i)}) - y^{(i)})x^{(i)}
$$

---

## 2. Stochastic Gradient Descent (SGD)
Updates weights for **each sample individually**. 
* *One epoch = $m$ weight updates* (where $m$ is the total number of samples).

$$
\theta = \theta - \alpha \cdot 2(h_\theta(x^{(i)}) - y^{(i)})x^{(i)}
$$

---

## 3. Mini-Batch Gradient Descent
Updates weights using **small batches** of samples. This is the sweet spot used in most modern deep learning.
* *One epoch = $\lceil m/b \rceil$ weight updates* (where $b$ is the batch size).

$$
\theta = \theta - \alpha \cdot \frac{2}{b} \sum_{i=1}^b (h_\theta(x^{(i)}) - y^{(i)})x^{(i)}
$$

---

## Comparison

| Method | Updates per Epoch | Stability | Speed (per epoch) |
| :--- | :--- | :--- | :--- |
| **Batch** | $1$ | High (smooth path) | Slow (computationally heavy) |
| **SGD** | $m$ | Low (noisy/erratic path) | Fast |
| **Mini-Batch** | $m/b$ | Medium | Medium (leverages vectorization) |

```python
import numpy as np

def gradient_descent(X, y, weights, learning_rate, n_epochs, batch_size=1, method='batch'):
    """
    Perform gradient descent optimization.
    
    Args:
        X: Feature matrix of shape (m, n)
        y: Target values of shape (m,)
        weights: Initial weights of shape (n,)
        learning_rate: Step size for gradient descent
        n_epochs: Number of complete passes through the dataset
        batch_size: Size of batches for mini-batch gradient descent (default: 1)
        method: Type of gradient descent ('batch', 'stochastic', or 'mini_batch')
    
    Returns:
        Optimized weights
    """
    m, n = X.shape
    optimized_weights = weights.copy()
    
    for _ in range(n_epochs):
        if method=='batch':
            # 1. Calculate predictions for all X
            pred = np.dot(X, optimized_weights)
            
            # 2. Calculate error
            errors = pred - y
            
            # 3. Calculate gradient: (2/m) * np.dot(X.T, error)
            gradient = (2/m) * np.dot(X.T, errors)
            
            # 4. Update optimized_weights
            optimized_weights -= learning_rate * gradient
            
        elif method == 'stochastic':
            for i in range(m):
                # 1. Get single sample
                X_i = X[i:i+1]
                y_i = y[i:i+1]
                
                # 2. Calculate predictions for X_i
                pred_Xi = np.dot(X_i, optimized_weights)
            
                # 3. Calculate error
                errors = pred_Xi - y_i
                
                # 4. Calculate gradient: (2/1) * np.dot(X_i.T, error)
                gradient = (2/1) * np.dot(X_i.T, errors)
                
                # 5. Update optimized_weights
                optimized_weights -= learning_rate * gradient
                
        elif method == 'mini_batch':
            for i in range(0, m, batch_size):
                # 1. Get batch
                X_batch = X[i : i+batch_size] 
                y_batch = y[i : i+batch_size]
                
                # 2. Find current chunk size
                k = len(X_batch)
                
                # 3. Calculate predictions for X_batch
                pred_Xbatch = np.dot(X_batch, optimized_weights)
                
                # 4. Calculate error
                errors = pred_Xbatch - y_batch
                 
                # 5. Calculate gradient: (2/k) * np.dot(X_batch.T, error)
                gradient = (2/k) * np.dot(X_batch.T, errors)
                
                # 6. Update optimized_weights
                optimized_weights -= learning_rate * gradient

    return optimized_weights
```

```python
# Test Case 1
import numpy as np

X = np.array([[1, 1], [2, 1], [3, 1], [4, 1]])
y = np.array([2, 3, 4, 5])
weights = np.zeros(X.shape[1])
learning_rate = 0.01
n_epochs = 100

# Test Batch Gradient Descent
output = gradient_descent(X, y, weights, learning_rate, n_epochs, method='batch')
print(output)

# Expected Output:
# [1.14905239 0.56176776]
```
