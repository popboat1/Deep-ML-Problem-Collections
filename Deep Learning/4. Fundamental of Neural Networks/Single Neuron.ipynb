# Single Neuron Model with Multidimensional Input and Sigmoid Activation

This task involves a **single neuron model** designed for **binary classification** with multidimensional input features, using the **sigmoid activation function** to output probabilities. It also involves calculating the **mean squared error (MSE)** to evaluate prediction accuracy.

---

## Mathematical Background

### Neuron Output Calculation

The neuron first computes a weighted sum of the input features and adds a bias:

$$
z = \sum_i (w_i \cdot x_i) + b
$$

where:

- $w_i$ = weight for feature $i$  
- $x_i$ = feature value  
- $b$ = bias term  

The result is then passed through the **sigmoid activation function**:

$$
\sigma(z) = \frac{1}{1 + e^{-z}}
$$

---

### MSE Calculation

The **Mean Squared Error (MSE)** measures the average squared difference between predicted values and true labels:

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (\text{predicted}_i - \text{true}_i)^2
$$

---

## Explanation of Terms

- $z$ — The **sum of weighted inputs plus bias**  
- $\sigma(z)$ — The **sigmoid activation output**  
- **predicted** — The **probabilities after sigmoid activation**  
- **true** — The **true binary labels**

---

## Practical Implementation

- Each **feature vector** is processed to compute a **weighted sum plus bias**.
- The result is passed through the **sigmoid function** to produce the **probability that the input belongs to the positive class**.
- The **MSE** is then calculated to measure the difference between predictions and the actual labels.

This error metric provides **insight into model performance** and helps guide **model optimization** during training.

[Problem Desc](https://www.deep-ml.com/problems/24?from=Deep%20Learning)

```python
import math

def single_neuron_model(features: list[list[float]], labels: list[int], weights: list[float], bias: float) -> (list[float], float):
    probabilities = []
    for row in features:
        z = sum(x_i * w_i for x_i, w_i in zip(row, weights)) + bias
        sigmoid_z = 1 / (1 + math.exp(-z))
        probabilities.append(round(sigmoid_z, 4))
        
    squared_errors = []
    for y, y_pred in zip(labels, probabilities):
        error = (y - y_pred)**2
        squared_errors.append(error)
        
    mse = sum(squared_errors) / len(labels)
    
    return probabilities, round(mse,4)
```

```python
# Test Case 1
print(single_neuron_model([[0.5, 1.0], [-1.5, -2.0], [2.0, 1.5]], [0, 1, 0], [0.7, -0.4], -0.1))

# Expected Output:
# ([0.4626, 0.4134, 0.6682], 0.3349)
```

```python
# Test Case 2
print(single_neuron_model([[1, 2], [2, 3], [3, 1]], [1, 0, 1], [0.5, -0.2], 0))

# Expected Output:
# ([0.525, 0.5987, 0.7858], 0.21)
```
