# Understanding Recurrent Neural Networks (RNNs)

**Recurrent Neural Networks (RNNs)** are a class of neural networks designed to handle **sequential data** by maintaining a **hidden state** that captures information from previous inputs.

---

## Mathematical Formulation

For each time step $t$, the RNN updates its hidden state $h_t$ using the current input $x_t$ and the previous hidden state $h_{t-1}$:

$$
h_t = \tanh(W_x x_t + W_h h_{t-1} + b)
$$

Where:

- $W_x$ is the **weight matrix for input-to-hidden connections**
- $W_h$ is the **weight matrix for hidden-to-hidden connections**
- $b$ is the **bias vector**
- $\tanh$ is the **hyperbolic tangent activation function** applied element-wise

---

## Implementation Steps

### 1. Initialization

Start with the **initial hidden state**:

$$
h_0
$$

---

### 2. Sequence Processing

For each input $x_t$ in the sequence:

$$
h_t = \tanh(W_x x_t + W_h h_{t-1} + b)
$$

The hidden state updates at every time step using both the **current input** and the **previous hidden state**.

---

### 3. Final Output

After processing all inputs, the **final hidden state**:

$$
h_T
$$

(where $T$ is the sequence length) contains information from the **entire sequence**.

---

## Example Calculation

Given:

- Inputs:  
  $x_1 = 1.0$, $x_2 = 2.0$, $x_3 = 3.0$

- Initial hidden state:  
  $h_0 = 0.0$

- Weights:  
  $W_x = 0.5$  
  $W_h = 0.8$

- Bias:  
  $b = 0.0$

---

### First Time Step ($t = 1$)

$$
h_1 = \tanh(0.5 \times 1.0 + 0.8 \times 0.0 + 0.0)
$$

$$
h_1 = \tanh(0.5) \approx 0.4621
$$

---

### Second Time Step ($t = 2$)

$$
h_2 = \tanh(0.5 \times 2.0 + 0.8 \times 0.4621 + 0.0)
$$

$$
h_2 = \tanh(1.0 + 0.3697) = \tanh(1.3697) \approx 0.8781
$$

---

### Third Time Step ($t = 3$)

$$
h_3 = \tanh(0.5 \times 3.0 + 0.8 \times 0.8781 + 0.0)
$$

$$
h_3 = \tanh(1.5 + 0.7025) = \tanh(2.2025) \approx 0.9750
$$

The final hidden state:

$$
h_3 \approx 0.9750
$$

---

## Applications

RNNs are widely used in tasks involving **sequential data**, such as:

- **Natural Language Processing (NLP)**
- **Time-series prediction**
- **Speech recognition**
- **Machine translation**
- **Text generation**

[Problem Desc](https://www.deep-ml.com/problems/54?from=Deep%20Learning)

```python
import numpy as np

def rnn_forward(
    input_sequence: list[list[float]], 
    initial_hidden_state: list[float], 
    Wx: list[list[float]], 
    Wh: list[list[float]], 
    b:list[float]) -> list[float]:
    
    x = np.array(input_sequence)
    W_x = np.array(Wx)
    W_h = np.array(Wh)
    h = np.array(initial_hidden_state)
    for i in range(len(input_sequence)):
        x_t = x[i]
        h = np.tanh(W_x @ x_t + W_h @ h + b)
    return np.round(h, 4)
```

```python
# Test Case 1
print(rnn_forward([[1.0], [2.0], [3.0]], [0.0], [[0.5]], [[0.8]], [0.0]))

# Expected Output:
# [0.9759]
```

```python
# Test Case 2
print(rnn_forward([[0.5], [0.1], [-0.2]], [0.0], [[1.0]], [[0.5]], [0.1]))

# Expected Output:
# [0.118]
```
