## Understanding the Softmax Activation Function

The **softmax function** is a generalization of the **sigmoid function** and is commonly used in the **output layer of neural networks for multi-class classification tasks**.

---

## Mathematical Definition

The softmax function converts a vector of scores (logits) into probabilities:

$$
\text{softmax}(z_i) = \frac{e^{z_i}}{\sum_j e^{z_j}}
$$

Where:

- $z_i$ is the score (logit) for class $i$
- The denominator sums the exponentials of all class scores

---

## Characteristics

### Output Range
Each output value lies between **0 and 1**, and the **sum of all outputs equals 1**:

$$
\sum_i \text{softmax}(z_i) = 1
$$

### Probability Distribution
Softmax transforms raw model outputs (logits) into a **probability distribution over classes**, making predictions interpretable and suitable for classification tasks.

---

## Numerical Stability

Directly computing $e^{z_i}$ can cause **numerical overflow** when input values are large.

To avoid this, we subtract the **maximum value of the vector** before exponentiation:

$$
\text{softmax}(z_i) =
\frac{e^{z_i - \max(z)}}{\sum_j e^{z_j - \max(z)}}
$$

This transformation is mathematically equivalent to the original expression because the factor $e^{-\max(z)}$ appears in both the numerator and denominator and therefore cancels out.

The benefit is that it keeps exponent values small and prevents overflow, making the implementation **numerically stable**.

---

## Why Softmax Is Important

The softmax function is essential for models where the output must represent a **probability distribution across multiple classes**, such as:

- Image classification
- Natural language processing tasks
- Multi-class prediction problems

It ensures that the model outputs **valid probabilities** that can be interpreted directly for decision making.

[Problem Desc](https://www.deep-ml.com/problems/23?from=Deep%20Learning)

```python
import math

def softmax(scores: list[float]) -> list[float]:
    z_max = max(scores)
    res = []
    for z_i in scores:
        temp = math.exp(z_i - z_max) / sum(math.exp(z_j - z_max) for z_j in scores)
        res.append(temp)
    return res
```

```python
# Test Case 1
print([round(x, 4) for x in softmax([1, 2, 3])])

# Expected Output:
# [0.09, 0.2447, 0.6652]
```

```python
# Test Case 2
print([round(x, 4) for x in softmax([1, 1, 1])])

# Expected Output:
# [0.3333, 0.3333, 0.3333]
```
